```python
train_df = train_df.drop(['POLICY_ID','VEHICLE_ID', 'CLAIM_PAID'], axis=1)
```

```python
train_df.info()
```

##### Carrying Capacity tiene un 23% de null values, los cuales podriamos reemplazar por el promedio, o eliminar de la base. Vamos a intentar primero removerlo del data frame. 

```python
train_df=train_df[train_df['CARRYING_CAPACITY'].notnull()]
```

## Data encoding

```python
#Transformo las variables de fechas a data type date
train_df['INSR_BEGIN'] = pd.to_datetime(train_df['INSR_BEGIN'], errors='coerce')
train_df['INSR_END'] = pd.to_datetime(train_df['INSR_END'], errors='coerce')
```

```python
train_df['START_YEAR'] = train_df['INSR_BEGIN'].dt.year
train_df['START_MONTH'] = train_df['INSR_BEGIN'].dt.month
train_df['START_DAY'] = train_df['INSR_BEGIN'].dt.day
train_df['END_YEAR'] = train_df['INSR_END'].dt.year
train_df['END_MONTH'] = train_df['INSR_END'].dt.month
train_df['END_DAY'] = train_df['INSR_END'].dt.day
```

```python
train_df = train_df.drop(['INSR_BEGIN','INSR_END'], axis=1)
```

```python
#Utilizo  label binarizer para asignar clasificacion binaria a la variable sexo
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
lb.fit(train_df['SEX'])
train_df['SEX'] = lb.transform(train_df['SEX'])
```

```python
#Para el caso de las categoricas Usage y Type of Vehicle utilizo one hot encoder
from sklearn.preprocessing import OneHotEncoder
encode= pd.get_dummies(train_df['USAGE'])
train_df=train_df.join(encode)
train_df=train_df.drop('USAGE', axis=1)
```

```python
encode= pd.get_dummies(train_df['TYPE_VEHICLE'])
train_df=train_df.join(encode)
train_df=train_df.drop('TYPE_VEHICLE', axis=1)
```

```python
#En el caso de Make como las categorias son mas de 600, y no quiero aumentar tanto la dimensionalidad del df, voy a utilizar hashing encoding
from sklearn.feature_extraction import FeatureHasher
train_df['MAKE'] = train_df['MAKE'].apply(lambda x: [x])

hasher = FeatureHasher(n_features=200).input_type='string')
hashed_features = hasher.transform(train_df['MAKE'].astype(str)).toarray()
```
